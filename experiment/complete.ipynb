```python
import pandas as pd
```

```python
df = pd.read_csv("not_fin3.csv")
```

```python
df["gloss"] = "-"
```

```python
df.head()
```

```python
with open("/home/hannusia/ucu/thesis/gloss_prompts/generated/glosses11.txt") as file:
    lines = file.readlines()

for line in lines:
    try:
        idd, lemma, gloss = line.split("\t")
        df.iloc[df[df["id_from"] == idd].index[0], 13] = gloss
    except:
        print(line)
```

```python
df.loc[:, "gloss"] = df["gloss"].apply(lambda x: "-" if x == "-. \n" else x)
```

```python
df["gloss"].value_counts()
```

```python
df.loc[:, "gloss"] = df["gloss"].apply(lambda x: "-" if (x == "яблуко - -. \n") or (x == "-\n") else x)
```

```python
df.fillna("-", inplace=True)
```

```python
entries = {}
```

```python
dfff = df[df["gloss"] != "-"].reset_index()
```

```python
for index, row in dfff.iterrows():
    # print(row)
    entries[row["lemma"]] = {"id_from": row["id_from"], "id_to": row["id_to"], "rel":row["rel"],
                                "ili": row["ili"], "lemma_en": row["lemma_en"], "gloss_en":row["gloss_en"], "gloss_uk": row["gloss"]}
```

```python
import json
with open("wn_with_glosses/wn_with_glosses6.json", "w") as outfile:
    json.dump(entries, outfile, indent=4, ensure_ascii=False)
```
