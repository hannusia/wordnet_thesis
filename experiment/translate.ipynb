```python
counter = 0
for index, row in df.iterrows():
    if not pd.notnull(row["lemma_en"]):
        try:
            synset = wn31.synset_from_pos_and_offset(row["id_crop"].split("-")[1], int(row["id_crop"][:-2]))
            row["lemma_en"] = str(synset)[8:-2].split('.')[0]
            row["gloss_en"] = synset.definition()
        except:
            id_crop = str(wn.synsets(ili=row["id_crop"].split("-")[-1])[1])[8:-2][-10:]
            synset = wn31.synset_from_pos_and_offset(id_crop.split("-")[1], int(id_crop[:-2]))
            row["lemma_en"] = str(synset)[8:-2].split('.')[0]
            row["gloss_en"] = synset.definition()
```
