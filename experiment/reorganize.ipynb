```python
import pandas as pd
import json
```

```python
df = pd.read_csv("merged/merged_mod.csv")
```

```python
df.shape
```

```python
df.head()
```

```python
df.fillna("", inplace=True)
```

```python
entries = []
ids = set()
```

```python
with open("wn_with_glosses/wn_with_glosses.json") as json_file:
    json_data = json.load(json_file)

for obj in json_data:
    if json_data[obj]["id_from"] not in ids:
        ids.add(json_data[obj]["id_from"])
        entry = {"lemma_uk": obj}
        entry.update(json_data[obj])
        entries.append(entry)

len(ids)
```
