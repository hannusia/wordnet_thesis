```python
import json
import pandas as pd
import ast
```

```python
from sentence_transformers import SentenceTransformer, util
model = SentenceTransformer('lang-uk/ukr-paraphrase-multilingual-mpnet-base')
```

```python
with open("wn_with_gloss.json", "r") as json_file:
    json_data = json.load(json_file)
```

```python
len(json_data)
```

```python
glosses = {}
for obj in json_data:
    try:
        if obj["lemma_uk"] in glosses:
            glosses[obj["lemma_uk"]].append(obj["gloss_uk"])
        else:
            glosses[obj["lemma_uk"]] = [obj["gloss_uk"]]
    except:
        print(obj)
        break
```

```python
llm_gloss = []
google_gloss = []
bing_gloss = []
deepl_gloss = []
gloss_en = []
for index, row in df_filter.iterrows():
    llm_gloss.append(glosses[row["lemma_ukr"]])
    google_gloss.append(df_tr[df_tr["lemma"] == row["lemma_ukr"]]["google_gloss"].to_list())
    bing_gloss.append(df_tr[df_tr["lemma"] == row["lemma_ukr"]]["bing_gloss"].to_list())
    deepl_gloss.append(df_tr[df_tr["lemma"] == row["lemma_ukr"]]["deepl_gloss"].to_list())
    gloss_en.append(df_tr[df_tr["lemma"] == row["lemma_ukr"]]["gloss_en"].to_list())

df_filter["llm_gloss"] = llm_gloss
df_filter["google_gloss"] = google_gloss
df_filter["bing_gloss"] = bing_gloss
df_filter["deepl_gloss"] = deepl_gloss
df_filter["gloss_en"] = gloss_en

df_filter
```

```python
lengths = df_filter["google_gloss"].apply(lambda x: len(x))
mask = lengths == 1

dff = df_filter[mask][["lemma_ukr", "gloss_sum", "llm_gloss", "google_gloss", "bing_gloss"]]
lengths = dff["gloss_sum"].apply(lambda x: len(x))
mask = lengths == 1
dff = dff[mask].reset_index(drop=True)

dff["gloss_sum"] = dff["gloss_sum"].apply(lambda x: str(x[0]))
dff["llm_gloss"] = dff["llm_gloss"].apply(lambda x: str(x[0]))
dff["google_gloss"] = dff["google_gloss"].apply(lambda x: str(x[0]))
dff["bing_gloss"] = dff["bing_gloss"].apply(lambda x: str(x[0]))
```

```python
dff["deepl_gloss"] = lines2
```

```python
dff
```

```python
def compare(option):
    if option == "google":
        definitions = dff["gloss_sum"].to_list()
        llm_gloss = dff["llm_gloss"].to_list()
        google_gloss = dff["google_gloss"].to_list()
        embeddings3 = model.encode(google_gloss, convert_to_tensor=True)
    elif option == "bing":
        definitions = dff["gloss_sum"].to_list()
        llm_gloss = dff["llm_gloss"].to_list()
        bing_gloss = dff["bing_gloss"].to_list()
        embeddings3 = model.encode(bing_gloss, convert_to_tensor=True)
    else:
        definitions = dff["gloss_sum"].to_list()
        llm_gloss = dff["llm_gloss"].to_list()
        deepl_gloss = dff["deepl_gloss"].to_list()
        embeddings3 = model.encode(deepl_gloss, convert_to_tensor=True)
    
    embeddings1 = model.encode(definitions, convert_to_tensor=True)
    embeddings2 = model.encode(llm_gloss, convert_to_tensor=True)

    cosine_scores_llm = util.cos_sim(embeddings1, embeddings2)
    cosine_scores_translator = util.cos_sim(embeddings1, embeddings3)

    scores_llm = []
    scores_translator = []
    for i in range(len(definitions)):
        scores_llm.append(round(float(cosine_scores_llm[i][i]), 4))
        scores_translator.append(round(float(cosine_scores_translator[i][i]), 4))
    
    return scores_llm, scores_translator
```

```python
scores_llm1, scores_google = compare("google")
scores_llm2, scores_bing = compare("bing")
scores_llm3, scores_deepl = compare("deepl")
```
